```python
env = UnityEnvironment(file_name="/Users/saboulaban/Documents/deep-reinforcement-learning/p1_navigation/Banana.app")
```

```python
brain_name = env.brain_names[0]
brain = env.brains[brain_name]
env_info = env.reset(train_mode=True)[brain_name]
```

```python
action_size = brain.vector_action_space_size
state_size = len(env_info.vector_observations[0])
```

```python
agent_duel_double_per = Duel_Double_PER_Agent(state_size=state_size, action_size=action_size, seed=42)
agent_duel_double = Duel_Double_Agent(state_size=state_size, action_size=action_size, seed=42)
agent_duel = Duel_Agent(state_size=state_size, action_size=action_size, seed=42)
agent_vanilla = Agent(state_size=state_size, action_size=action_size, seed=42)
```

```python
agent_vanilla.qnetwork_local.load_state_dict(torch.load('../models/vanilla.dqn.m-7.pth'))
agent_duel.qnetwork_local.load_state_dict(torch.load('../models/duel.dqn.m-7.pth'))
agent_duel_double.qnetwork_local.load_state_dict(torch.load('../models/duel.ddqn.m-8.pth'))
agent_duel_double_per.qnetwork_local.load_state_dict(torch.load('../models/duel.ddqn.per.m-9.pth'))
```

```python
agents = {
    'vanilla_agent': agent_vanilla,
    'agent_duel': agent_duel,
    'agent_duel_double': agent_duel_double,
    'agent_duel_double_per': agent_duel_double_per
}
for agent_name, agent in agents.items():
    agents_scores = {}
    scores = []
    for i in range(100):
        env_info = env.reset(train_mode=True)[brain_name]
        state = env_info.vector_observations[0]
        score = 0
        for j in range(1000):
            action = agent.act(state, 0.1)
            env_info = env.step(action)[brain_name]
            state = env_info.vector_observations[0]
            reward = env_info.rewards[0]
            done = env_info.local_done[0]
            score += reward
            if done:
                scores.append(score)
                break
        scores.append(score)
    print("100 epsiodes average score for ", agent_name, " is : ", np.mean(scores))
    agents_scores[agent] = np.mean(scores)
```

```python
#Best performing agent playing the game is Duel DDQN PER Agent
env.close()
```
