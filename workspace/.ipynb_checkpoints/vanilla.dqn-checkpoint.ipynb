```python
class Helper:
    def LiveUpdatePlot(self, x, y):
        plt.ion()
        fig = plt.figure()
        ax = fig.add_subplot(111)
        line1, = ax.plot(x, y, 'b-')
        return line1
```

```python
from collections import deque
class RandomAgent:
    def __init__(self, env, brain, brain_name, episodes_resuls_size = 100, num_episodes = 500, print_every = 10, plot_every = None):
        self.env = env
        self.brain = brain
        self.brain_name = brain_name
        self.env_info = self.env.reset(train_mode=True)[brain_name]
        self.episodes_resuls_size = episodes_resuls_size
        self.num_episodes = num_episodes
        self.action_size = self.brain.vector_action_space_size
        self.state_size = len(self.env_info.vector_observations[0])
        self.episodes_results = deque(maxlen = self.episodes_resuls_size)
        self.results = []
        self.print_every = print_every
        self.plot_every = plot_every
    def train_agent(self):
        for episode in range(self.num_episodes):
            env_info = self.env.reset(train_mode=True)[self.brain_name]
            state = env_info.vector_observations[0]
            score = 0
            while True:
                action = np.random.randint(self.action_size)
                env_info = self.env.step(action)[self.brain_name]
                next_state = env_info.vector_observations[0]
                reward = env_info.rewards[0]
                done = env_info.local_done[0]
                score += reward
                state = next_state
                if done:
                    self.episodes_results.append(score)
                    break
            if len(self.episodes_results) == self.episodes_resuls_size:
                self.results.append(np.asarray(self.episodes_results).mean())
            if episode % self.print_every == 0 and len(self.results) >= self.episodes_resuls_size:
                print("Average score for last n episodes is: ", np.asarray(self.episodes_results).mean())
            if self.plot_every is not None:
                if episode % self.plot_every == 0:
                    plt.plot(self.results)
                    plt.show()
        plt.plot(self.results)
        plt.show()

        
                    
```

```python
env = UnityEnvironment(file_name="/Users/saboulaban/Documents/deep-reinforcement-learning/p1_navigation/Banana.app")
```

```python
brain_name = env.brain_names[0]
brain = env.brains[brain_name]
```

```python
randomAgent = RandomAgent(env, brain, brain_name, num_episodes = 1800, plot_every = 300, print_every = 10)
```

```python
randomAgent.train_agent()
```
